```python
import matplotlib.pyplot as plt
import numpy as np 
from scipy import stats 
import pandas as pd 
import seaborn as sns  

datos = pd.read_csv('Sopas_en_lata.csv',delimiter=',')

datos.dtypes

datos.describe()
```

```python
plt.figure()
plt.hist(datos['COSTO'], bins = 5, edgecolor = 'black',  linewidth=1)
plt.xlabel('COSTO')
plt.ylabel('Frecuencia')
plt.show()
sns.histplot(data=datos['COSTO'], bins=5, kde=True)
plt.show()
```
